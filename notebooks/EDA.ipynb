```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
fashion_mnist = keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

```python
fig, ax = plt.subplots(len(np.unique(train_labels)), 5, figsize=(15,15))
for i in range(0, len(np.unique(train_labels))):
    train_samples = train_images[np.where(train_labels == i)[0][:5]]
    for j in range(train_samples.shape[0]):
        ax[i, j].imshow(train_samples[j], cmap=plt.cm.binary)
```
